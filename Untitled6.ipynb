```python

# Put all frames you want to change the flag IMG_TYP to object in a directory (say light) and
# then run the following. Repeat for bias, dark or flat frames making sure you change the keyword 'object' to the 
# correct file type you want each time.

reduced_path = Path(r'C:/Users/sarin/OneDrive/Desktop/Rezaei_Dec01_2024/Darks/dark_5min_-25C')

# Extract the filenames of the image files
image_files = [os.path.join(reduced_path, filename) for filename in os.listdir(reduced_path) if filename.endswith(".fit")]

for image_file in image_files:
    with fits.open(image_file, mode='update') as hdulist:
        header = hdulist[0].header
        header['IMG_TYP'] = 'dark'

        if 'BUNIT' not in header:
            header['BUNIT'] = 'adu'

        if 'IMG_TYP' not in header:
            header['IMG_TYP'] = 'dark'

        hdulist.flush()  # Saves changes automatically

hdulist.close()  # Close the file
image=fits.open(r'C:\Users\sarin\OneDrive\Desktop\Rezaei_Dec01_2024\Darks\dark_5min_-25C\dark_bin2_-25C_2024_12_01_2x2_exp00.05.00.000_Low_3.fit')
image[0].header
```
